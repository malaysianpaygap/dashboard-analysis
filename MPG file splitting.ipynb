```python
# function to create dim dataframe/table
def create_dim(col_name, col_head):
    col_en_name = col_head + '_en'
    col_bm_name = col_head + '_bm'
    col_key  = col_head + '_key'
    
    df_dim = df_main[col_name].to_frame()
    
    # rename en col
    df_dim.rename(columns = {col_name : col_en_name}, inplace = True)
    # drop duplicates & na
    df_dim.drop_duplicates(inplace = True)
    df_dim.dropna(inplace = True)
    # add bm col
    df_dim.insert(loc = 1, column = col_bm_name, value = None)
    # sort alphabetically
    df_dim.sort_values(by = [col_en_name], ascending = True, inplace = True)
    
    # reset index
    df_dim.reset_index(drop = True, inplace = True)
    # add key col
    df_dim[col_key] = df_dim.index + 1
    # rearrange cols
    cols = df_dim.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df_dim = df_dim[cols]
    
    return df_dim    
```

```python
# fuction to save dataframe as csv dim
def save_dim_csv(df, col_head):
    csv_name = r'C:\Users\Dante\Desktop\MPG\CSV files\dim_' + col_head + '.csv'
    df.to_csv(csv_name, index = False)
```

```python
# function to extract dim table, replace values, rename column
def replace_dim(col_name, col_head):
    
    # extract dim table
    df_dim = pd.read_csv(r'C:\Users\Dante\Desktop\MPG\CSV files\dim_' + col_head + '.csv')
    
    # assign names
    col_en_name = col_head + '_en'
    col_key  = col_head + '_key'

    # convert dim table into dict
    df_dim = df_dim[[col_key, col_en_name]]
    dim_dict = df_dim.set_index(col_en_name)[col_key].to_dict()
        
    # replace name with key
    for key, value in dim_dict.items():
        df_main.replace(to_replace = key, value = value, inplace = True)
        
    # rename col
    df_main.rename(columns = {col_name : col_key}, inplace = True)
```
